2.1) Έστω $f(x) = x^T Ax + b^T x$, μια συνάρτηση d μεταβλητών, με Α συμμετρικό πίνακα (δηλ $A = A^T$) διάστασης $dxd$, και $b \in R^d$. Να υπολογίσετε αναλυτικά βήμα προς βήμα με το χέρι, την παράγωγο της $f(x)$ ως προς το διάστημα $x \in R^d$ παραθέτοντας όλα τα ενδιάμεσα βήματα:

Από τις ιδιότητες των Derivatives του Matrix Cookbook, γνωρίζουμε ότι ισχύει: <br/> <center> ${\partial (X + Y)} = {\partial X} + {\partial Y}$.</center> 
Άρα η παράγωγος της **f**, είναι ίση με το άθροισμα των παραγώγων δύο άλλων συναρτήσεων, έστω **g** και **h**: ${\partial f(x)} = {\partial (x^TAx + b^Tx} = {\partial (x^TAx)} + {\partial (b^Tx)} = {\partial g(x)} + {\partial h(x)}$. Με $g(x) = x^TAx$ και $h(x) = b^Tx$ <br/>
Άρα βρίσκοντας τις παραγώγους των g και h ξεχωριστά και αθροίζοντάς τες θα βρούμε την παράγωγο της f:


<h3> i) Παράγωγος της $g(x) = x^TAx$:

Έχουμε από υπόθεση ότι $x \in R^d$ και $A \in R^{dxd}$. <br/>
Από τον **ορισμό πολλαπλασιασμού πινακα με διανυσμα** και τον **ορισμό εσωτερικού γινομένου**: <br/><br/>

$g(x) = x^TAx = x^T \begin{bmatrix} 
\sum_{j=1}^d A_{1j} x_j \\
\sum_{j=1}^d A_{2j} x_j \\
... \\
\sum_{j=1}^d A_{dj} x_j 
\end{bmatrix}
= \sum_{i=1}^d \sum_{j=1}^d x_i A_{ij} x_j$<br/><br/>

Παραγωγίζοντας έχουμε: $ \frac{\partial g(x)}{\partial x} = \frac{\partial}{\partial x_k} \sum_{i=1}^d \sum_{j=1}^d x_i A_{ij} x_j =$ <br/> $ = \sum_{i=1}^d \sum_{j=1}^d \frac{\partial}{\partial x_k} (A_{ij} x_i x_j) $ (Αφού $A_{ij} x_i x_j = x_i A_{ij} x_j$)<br/>
$ = \sum_{i=1}^d \sum_{j=1}^d A_{ij} \frac{\partial}{\partial x_k} (x_i x_j)
$ (Αφού $A_{ij} \in R$) **(σχέση 1)** <br/><br/> 
Από εδώ, Θα πάρουμε τις εξής περιπτώσεις: <br/>
1) $(k != i) and (k != j) => \frac{\partial}{\partial x_k} (x_i x_j) = 0$<br/>
2) $k = i => \frac{\partial}{\partial x_k} (x_i x_j) = x_j$<br/>
3) $k = j => \frac{\partial}{\partial x_k} (x_i x_j) = x_i$<br/><br/>

Συνεπώς, τελικά η **(σχέση 1)** θα γίνει ίση με το άθροισμα μόνον όσων στοιχείων της <br/> δεν μηδενίζονται με βάση τις παραπάνω περιπτώσεις, δηλαδή:<br/>
(σχέση 1) $= \sum_{i=1}^d A_{ik} x_i + \sum_{j=1}^d A_{kj} x_j
$**(σχέση 2)**<br/><br/>

Η **(σχέση 2)**, όπως είδαμε και στο ερώτημα 1.1 και στις διαφάνειες του μαθήματος (L02 - slide 11), αντιστοιχεί στο άθροισμα δύο εσωτερικών γινομένων: <br/><br/>
i) Το άθροισμα $\sum_{i=1}^d A_{ik} x_i$ αντιστοιχεί στο **εσωτερικό γινόμενο της ανάστροφης κ-οστης στήλης** του πίνακα Α με το διάνυσμα x, και <br/>
ii) Το άθροισμα $\sum_{j=1}^d A_{kj} x_j$ αντιστοιχεί στο **εσωτερικό γινόμενο της κ-οστης σειράς** του πίνακα Α με το διάνυσμα x<br/><br/>

Αρα τελικά από τον γνωστό ορισμό του εσωτερικού γινομένου και την τελευταία παρατήρηση, η **(σχέση 2)** γίνεται:
**(σχέση 2)** **$= A^Tx + Ax = Ax + Ax = 2Ax$** <br/> Εφόσον από υπόθεση ισχύει $A^T=A$ αφού ο πίνακας Α είναι συμμετρικός.<br/><br/>

**Τελικά $\triangledown g(x) = 2Ax$**


<h3> ii) Παράγωγος της $h(x) = b^Tx$:

$h(x) = b^Tx = \sum_{i=1}^d b_{i} x_{i}$ (Από ορισμό Εσωτερικού γινομένου διανυσμάτων) <br/>

Από την υπενθύμιση της εκφώνησης, η παράγωγος της συνάρτησης h θα είναι: <br/>
<br/>
$\triangledown h(x) = 
\begin{bmatrix}
\frac{\partial h(x)}{\partial x_1} \\
... \\
\frac{\partial h(x)}{\partial x_d}
\end{bmatrix} =
\begin{bmatrix}
b_1 \\
b_2 \\
... \\
b_d
\end{bmatrix} = b
$<br/><br/>

Αφού ισχύει:
$\frac{\partial h(x)}{\partial x_1} = \frac{\partial (b_1x_1)}{\partial x_1} = b_1$ <br/>(όμοια για i= 2, 3, .. d, αφού για κάθε i, μονο το $x_i$ και ο συντελεστής του δεν μηδενίζουν) <br/><br/>

Τελικά: $\triangledown h(x) = b$




Από τα i), ii) και από την αρχική πρόταση, έχουμε: <br/>
$\triangledown f(x) = \triangledown g(x) + \triangledown h(x)$, δηλαδή <br/>
**$\triangledown f(x) = 2Ax + b$**<br/>
Η οποία λύση, για τον Α συμμετρικό $(A = A^T)$, <br/>είναι η ίδια με τον γενικό τύπο
 που δίνει το matrix cookbook για τα gradients (σελ 12), οπότε και επαληθεύεται.

2.2) Έστω Α, Β, Χ πίνακες πραγματικών αριθμών διάστασης d x d με τάξη (rank) ίση με d. Να βρείτε, αναλυτικά και σε κλειστή μορφή (closed form solution) το ολικό ελάχιστο του παρακάτω προβλήματος βελτιστοποίησης:
<center>$\min_{x} ||A-XB||_{F}^{2}$</center>
Να εξηγήσετε με λεπτομέρεια όλη τη διαδικασία σκέψης που θα ακολουθήσετε και να παραθέσετε όλα τα ενδιάμεσα βήματα των πράξεων:

Στόχος μας είναι να λύσουμε ενα πρόβλημα ελαχιστοποίησης. Όπως είδαμε στη θεωρία, ένας τρόπος να λυθούν τέτοια προβλήματα είναι η εύρεση του ελαχίστου μέσω της παραγώγου της σχέσης μας, στην περίπτωση που είναι **κυρτή**.

Στην περίπτωση μας, η ποσότητα που θέλουμε να ελαχιστοποιήσουμε είναι όντως κυρτή, καθώς είναι μια Frobenius Νόρμα (οι νόρμες είναι κυτρές) ή οποία είναι υψωμένη στο τετράγωνο (επίσης κυρτή). Άρα, μπορούμε να βρούμε το ελάχιστο της μέσω παραγώγισης:

Πρέπει να βρούμε σε ποιό σημείο μηδενίζεται η πρώτη παράγωγος της εξής συνάρτησης: <br/> Έστω  $f(X) = \||A-XB||_{F}^{2}$ <br/> 

Στο σημείο αυτό, παραθέτω τις ιδιότητες που χρησιμοποιώ στις πράξεις μου παρακάτω, αριθμημένες για μια πιο καθαρή απόδειξη: <br/>( Oι αριθμοί που ακολουθούν κάθε ιδιότητα αντιστοιχούν στους αριθμούς σχέσεων στο "The Matrix Cookbook" - **TMC** )<br/> <br/>

(1) Derivative of Frobenius Norm<br/>
$\frac{\partial}{\partial X} ||X||_{F}^{2} = \frac{\partial}{\partial X} Tr(XX^T)$ : (TMC 132)<br/> <br/>

(2) ${\partial (X+Y) } = ({\partial X} + {\partial Y})$ : (TMC 35)<br/> <br/>

(3) $(A + B)^T = A^T + B^T$ : (TMC 4)<br/> <br/>

(4) $(AB)^T = B^TA^T$ : (TMC 5)<br/> <br/>

(5) $ Tr(A+B) = Tr(A) + Tr(B)$ : (TMC 15)<br/> <br/>

(6) $ Tr(AB) = Tr(BA)$ : (TMC 14)<br/> <br/>

και αντίστοιχα: $ Tr(ABC) = Tr(BCA) = Tr(CAB)$ : (TMC 16)<br/> <br/>

(7) $ \frac{\partial}{\partial X} Tr(AXB) = A^TB^T$ : (TMC 101)<br/> <br/>

(8) $ \frac{\partial}{\partial X} Tr(AX^TB) = BA$ : (TMC 102)<br/> <br/>

(9) $ \frac{\partial}{\partial X} Tr(XBX^T) = XB^T + XB$ : (TMC 111)<br/> <br/>

(10) $ \frac{\partial}{\partial X} Tr(XBB^TX^T) = Tr(X (BB^T) X^T) =^{(9)} X(BB^T)^T + X(BB^T) = XBB^T + XBB^T = 2XBB^T$

$f(X) = \||A-XB||_{F}^{2} =>$ <br/>
$\frac{\partial f}{\partial X} =^{(1)}
\frac{\partial}{\partial X} Tr[ (A-XB)(A-XB)^T ] =^{(3)}
\frac{\partial}{\partial X} Tr[ (A-XB)(A^T-(XB)^T) ] =$ <br/>
$=^{(4)} 
\frac{\partial}{\partial X} Tr[ (A-XB)(A^T-B^TX^T) ] =
\frac{\partial}{\partial X}Tr[ AA^T - AB^TX^T - XBA^T + XBB^TX^T] =$ <br/>
$=^{(5)} 
\frac{\partial}{\partial X} [ Tr(AA^T) - Tr(AB^TX^T) - Tr(XBA^T) + Tr(XBB^TX^T)] =$ <br/> $=^{(6)}
\frac{\partial}{\partial X} [ Tr(AA^T) - Tr(B^TX^TΑ) - Tr(Α^ΤXΒ) + Tr(XBB^TX^T)]$ <br/>
$ =^{(7),(8),(10)} 0 - AB^T - AB^T + 2XBB^T = 2XBB^T - 2AB^T$ <br/>
( Στο (6), επέλεξα να αφήσω το $B^T$ δεύτερο καθώς επιτρέπει απλοποιήσεις παρακάτω στις πραξεις )

Βρήκαμε ότι $\triangledown f(x) = 2XBB^T - 2ΑB^T$, και όπως ειδαμε παραπάνω η f είναι κυρτή συνάρτηση, άρα για το ελάχιστό της αρκεί να βρούμε που μηδενίζει η παράγωγος:

$\triangledown f(x) = 0 =>  2XBB^T - 2AB^T = 0 => \\
2XBB^T = 2AB^T =>$ <br/> 
$XBB^T = AB^T =>$ (Πολλαπλασιάζουμε με τους κατάλληλους αντίστροφους πίνακες ώστε να απομονώσουμε το X)
$XBB^T(B^T)^{-1} = AB^T(B^T)^{-1} =>$ <br/> 
$XBB^{-1} = AB^{-1} =>$ <br/> 
$X = AB^{-1}$

Άρα τελικά, για $X = AB^{-1}$, η παράγωγος της f έχει ολικό ελάχιστο αφού είναι κυρτή, άρα λύσαμε το ζητούμενο πρόβλημα βελτιστοποίησης.

# Ερώτημα 3: Gradient descent
---

Ο αλγόριθμος gradient descent είναι η πιο διαδεδομένη μέθοδος βελτιστοποίησης στη μηχανική μάθηση.

Σε αυτό το ερώτημα καλείστε να υλοποιήσετε σε Python τον αλγόριθμο
**gradient descent** για την ελαχιστοποίηση συναρτήσεων δύο μεταβλητών.<br/> 
**Θεωρήστε τις παρακάτω συναρτήσεις:** <br/>
> Nα υλοποιήσετε σε Python και να εκτελέστε τον αλγόριθμο καθόδου κλίσης (gradient descent) για κάθε συνάρτηση.
* $f_1([x_1,x_2]^T) = f_1(x_1,x_2) = (x_1 - 2)^2 + (x_2 - 3)^2$ 
* $f_2([x_1,x_2]^T) = f_2(x_1,x_2) = (1 -(x_2 - 3))^2 + 20((x_1+3) - (x_2-3)^2 )^2$

```python
# For simplicity, i use symbols (like we did in matlab) for the functions
import sympy as sym
from sympy import lambdify
x = sym.symbols('x')
y = sym.symbols('y')
```

```python
# A function that given a symbolic 2-argument function, returns a tuple with
# both itself and its derivative as lambda functions
def derivativeOf(f):
  # Calculate the partial derivatives
  dfx = sym.diff(f, x)
  dfy = sym.diff(f, y)

  # Convert f1 to a lambda function
  fl = lambdify((x,y),f)
  # Same with the derivatives
  dfx = lambdify((x,y),dfx)
  dfy = lambdify((x,y),dfy)

  # Combine both partial derivatives into a single function
  df = lambda x, y: (dfx(x,y), dfy(x,y))
  return (fl, df)
```

```python
# Start with f1 and f2 in a symbolic form
f1 = (x-2)**2 + (y-3)**2
f2 = (1-(y - 3))**2 + 20*((x+3)-(y-3)**2)**2

# Get f1 and its derivative as lambda functions
f1, df1 = derivativeOf(f1)
# Same for f2
f2, df2 = derivativeOf(f2)
```

```python
# The gradient descend variable itself (For 2-argument functions only)
def gradient_descend(f, df, x, y, rate, verbose):
  iteration_vals = [] # Add the value of the function after each iteration in a list to also return them
  if verbose == 1:
    print(" (i) Starting from x=%d, y=%d" % (x,y))
  iterations = 1
  while iterations < (max_iterations+1):
    prevx = x
    prevy = y
    try: # Handle possible overflow exceptions
      tempx, tempy = df(x, y)
      x = x - rate*tempx
      y = y - rate*tempy
      fval = f(x, y)
      if verbose == 1:
        print("     Iteration %3d: f(%.2f, %.2f) = %.6f" % (iterations, x, y, fval))
      iteration_vals.append(fval)

      if (abs(df(x,y)[0]) <= tolerance) and (abs(df(x,y)[1]) <= tolerance):
        if verbose == 1:
          print(" (!) Solution found: x=%.3f, y=%.3f" % (x,y))
          print("     Total iterations: %d\n" % iterations)
        return (iteration_vals, x, y, iterations)

    except Exception:
      pass
      if verbose == 1:
        print(" (!) No Solution Found due to overflow after %d iterations" % (iterations-1))
      return (iteration_vals, None, None, None)

    iterations += 1

  if verbose == 1:
    print(" (!) Maximum amount of iterations reached without convergence, stopping..." )
  return (iteration_vals, x, y, None)
```
